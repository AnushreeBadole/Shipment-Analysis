```python
# Calculate 'Cost per km'
df['Cost per km'] = df['Shipment Cost (in Rs.)'] / df['Distance (km)']
```

```python
# Calculate 'Cost per kg'
df['Cost per kg'] = df['Shipment Cost (in Rs.)'] / df['Package Weight (kg)']
```

```python
# Create 'Delay Category'
def categorize_delay(delay):
    if delay <= 0:
        return 'On Time'
    elif 0 < delay <= 24:
        return 'Low'
    elif 24 < delay <= 48:
        return 'Medium'
    else: # delay > 48
        return 'High'
```

```python
df['Delay Category'] = df['Delay (hrs)'].apply(categorize_delay)
```

```python
# Create 'Is Delayed' boolean feature
df['Is Delayed'] = df['Delay (hrs)'] > 0

print("\nFirst 5 rows with new features:")
print(df[['Shipment ID', 'Cost per km', 'Cost per kg', 'Delay Category', 'Is Delayed']].head())
```

```python
high_val_threshold = df['Package Cost (in Rs.)'].quantile(0.75)
high_val_threshold
```

```python
df['Is Fragile & High Value'] = (df['Fragile'] == True) & (df['Package Cost (in Rs.)']> high_val_threshold)
```

```python
df.head()
```

```python
q1 = df['Total Cost (in Rs.)'].quantile(0.25)
q3 = df['Total Cost (in Rs.)'].quantile(0.75)
```

```python
IQR = q3-q1
print(q1)
print(q3)
print(IQR)
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
df['Outliered'] = (df['Total Cost (in Rs.)'] < lower_bound) | (df['Total Cost (in Rs.)'] > upper_bound)
```

```python
df['Total Cost Capped'] = np.where(df['Total Cost (in Rs.)'] < lower_bound, lower_bound,
    np.where(df['Total Cost (in Rs.)'] > upper_bound, upper_bound, df['Total Cost (in Rs.)']))
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Total Cost Capped'],kde=True)
plt.title("Total Cost after outlier removal")
plt.show()
```

**ENCODING AND SCALING**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = ['Company Name', 'Shipment Priority', 'Shipment Mode', 'Status', 'Payment Method']
df_cat = df[categorical_cols]
```

```python
ohh = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')
ohh_data = ohh.fit_transform(df_cat)
```

```python
ohh.get_feature_names_out(categorical_cols)
```

```python
encoded_df = pd.DataFrame(ohh_data, columns=ohh.get_feature_names_out(categorical_cols))
```

```python
df_encoded = df.drop(columns=categorical_cols)
df_final_encoded = pd.concat([df_encoded, encoded_df], axis=1)
```
